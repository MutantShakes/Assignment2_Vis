```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Open the .nc file
file_path = '8b5f1222ef03cc9bf468a0dee16ddcd4_ferret_listing.nc'
nc_file = nc.Dataset(file_path, 'r')

# Access the sea_ice_mask variable
sea_ice_mask_data = nc_file.variables['sea_ice_mask'][:, :, :]

# Access latitude and longitude variables
lat = nc_file.variables['lat'][:]
lon = nc_file.variables['lon'][:]

# Close the .nc file to release resources
nc_file.close()

# Experiment with different colormaps
colormaps = ['viridis', 'coolwarm', 'cividis', 'magma', 'jet']
```

```python
# Experiment with parametric mapping using global vs local maxima and minima
min_value = sea_ice_mask_data.min()
max_value = sea_ice_mask_data.max()
```

```python
for time_step in range(sea_ice_mask_data.shape[0]):
    sea_ice_mask_data_timestep = sea_ice_mask_data[time_step, :, :]

    # Plot using a diverging colormap with local minima and maxima for each timestep
    plt.imshow(sea_ice_mask_data_timestep, cmap='RdBu_r', vmin=sea_ice_mask_data_timestep.min(), vmax=sea_ice_mask_data_timestep.max(), extent=(lon.min(), lon.max(), lat.min(), lat.max()))
    plt.colorbar(label='Sea Ice Mask')
    plt.title(f'Sea Ice Mask Distribution - Timestep: {time_step}')
    plt.xlabel('Longitude (degrees_east)')
    plt.ylabel('Latitude (degrees_north)')
    plt.show()

    # Plot using a continuous colormap with global minima and maxima across all timesteps
    plt.imshow(sea_ice_mask_data_timestep, cmap='viridis', vmin=min_value, vmax=max_value, extent=(lon.min(), lon.max(), lat.min(), lat.max()))
    plt.colorbar(label='Sea Ice Mask')
    plt.title(f'Sea Ice Mask Distribution - Timestep: {time_step}')
    plt.xlabel('Longitude (degrees_east)')
    plt.ylabel('Latitude (degrees_north)')
    plt.show()
```
